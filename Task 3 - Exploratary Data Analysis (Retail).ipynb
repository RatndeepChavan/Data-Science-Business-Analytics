```python
df = df.drop('Country', axis=1)
```

```python
pf = df[df['Profit'] > 0]
ls = df[df['Profit'] < 0]
no = df[df['Profit'] == 0]

print('Number observation where no profit as well as loss occure : '+ str(len(no)))

for i in df.columns:
    if df[i].nunique() < 20:
        plt.figure(figsize=(17,5))
        ax = sns.countplot(df[i])
        
        v = list(df[i].unique())
        if pd.api.types.is_numeric_dtype(df[i]):
            v.sort()
        
        c = 0
        for p in ax.patches:
            try : 
                c1 = df[i].value_counts()[v[c]]
                txt = str(c1) + '\n( '+ str(round(c1/len(df)*100,1))+ '%)\n'
            except:
                pass
            
            try : 
                c2 = pf[i].value_counts()[v[c]]
                txt+= 'P=' + str(int((c2/c1)*100)) + '%\n'
            except:
                pass
            
            try:
                c3 = ls[i].value_counts()[v[c]]
                txt+= 'L='+str(int((c3/c1)*100)) + '%\n'
            except:
                pass
            
            c+= 1
            
            ax.text(p.get_x()+p.get_width()/2, p.get_height(), txt, fontsize=12, ha='center')
            
        txt2 = 'Total orders\n'
        txt2+= 'Percent of contribution in total orders\n'
        txt2+= 'Percent of profit orders out of total orders for that category\n'
        txt2+= 'Percent of loss orders out of total orders for that category'
        ax.text(0, -0.6, txt2, fontsize=14, ha='left', transform=ax.transAxes)
            
        plt.show()
        
        
        plt.figure(figsize=(17,5))
        ax = sns.barplot(x=i, y='Profit', data=df, ci=None)
        
        v = list(df[i].unique())
        c = 0
        for p in ax.patches:
            l1 = pf[pf[i]==v[c]]['Profit']
            l2 = ls[ls[i]==v[c]]['Profit']
            t1 = (l1.sum()/pf['Profit'].sum())*100
            t2 = (l2.sum()/ls['Profit'].sum())*100
    
            txt = str(int(p.get_height())) + '\n'
            txt+= 'P=' + str(int(t1)) + '%\n' 
            txt+= 'L=' + str(int(t2)) + '%\n'
            
            ax.text(p.get_x()+p.get_width()/2, p.get_height(), txt, fontsize=12, ha='center')
            c += 1
            
        txt2 = 'Average profit including profit and loss\n'
        txt2+= 'Percent of contribution in total profit\n'
        txt2+= 'Percent of contribution in total loss'
        ax.text(0, -0.6, txt2, fontsize=14, ha='left', transform=ax.transAxes)
        plt.show()
```
